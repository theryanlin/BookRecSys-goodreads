```python
import pandas as pd
import numpy as np
from PIL import Image
import requests

ratings = pd.read_csv('/content/gdrive/MyDrive/RecSys Project 2/goodbooks-10k-dataset/ratings.csv')  
books = pd.read_csv('/content/gdrive/MyDrive/RecSys Project 2/goodbooks-10k-dataset/books.csv')
```

```python
def get_img(url):
  size = (224,224)
  img = Image.open(requests.get(url, stream=True).raw)
  img = img.resize(size,Image.ANTIALIAS)
  try:
    return np.array(img.getdata()).reshape(img.size[0], img.size[1], 3)
  except:
    img = img.convert("RGB")
    return np.array(img.getdata()).reshape(img.size[0], img.size[1], 3)
```

```python
from tqdm import tqdm
tqdm.pandas(desc='pandas bar')
```

```python
temp1 = books.image_url.progress_apply(get_img)
```

```python
temp1 = temp1.values
# np.save('pics6', temp1)
```

```python
n_train = 10000
train_X = np.asarray([temp1[i].reshape((-1,1)) for i in range(n_train)])
X = train_X.reshape((n_train,224,224,3))
```

```python
codings1 = model2.predict(X)
```
